```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import datetime
import os
pd.set_option('display.max_columns', None)
import sqlite3
from sqlalchemy import create_engine
```

```python
url = "https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/2019"
```

```python
dados = requests.get(url)
```

```python
soup = BeautifulSoup(dados.content, 'html.parser')
```

```python
print(soup.title)
```

```python
table = soup.findAll('table', attrs="table border-body")[0]
```

```python
colunas = []
for thead in table.findAll('thead'):
    #print(thead.getText())
    colunas.append(thead.getText())
    break
```

```python
tbody = table.findAll('tbody')[0]
```

```python
linhas = []
for tr in tbody.findAll('tr'):
    img = tr.find('img')
    print('Clube:', img['title'][:img['title'].find('-')-1])
    
    td = tr.findAll('td')
    print('Nome:', td[1].getText())
    #
    print('Apelido:', td[2].getText())
    #
    th = tr.findAll('th')
    print('Gols:', th[0].getText())
    #
    tupla = (img['title'][:img['title'].find('-')-1], td[1].getText(), td[2].getText(), int(th[0].getText()))
    linhas.append(tupla)
```

```python
linhas
```

```python
df = pd.DataFrame(linhas, columns=['Clube', 'Nome', 'Apelido', 'Gols'])
```
